Check if Dataset is Null or Empty

```python
def load_file(path): # path is new_data
    try:
        data = pd.read_csv(path)

        if data.isnull().sum() != 0:
            data.dropna()
            print("null values in the dataset were dropped\n")
        elif not data.sum() : #  elif not data.sum() 
            data= data.replace(r'^\s*$',float('NaN'),regex = True)
            data.isnull().sum() != 0
            data.dropna()
            print("empty values are converted to null values and were droped from the dataset\n")
        else:
            print("their is no null in the dataset\n")

    except:
        print('There is an Error!\n')

    return data    
```

Change Values


```python
def change_value(data):
    old_value = 5.5
    new_value = 4.2

    setosa = 'Iris-setosa'
    new_setosa = 1

    virginica = 'Iris-virginica'
    new_virginica = 2

    versicolor = 'Iris-versicolor'
    new_versicolor = 3

    data['sepalLengthInput'] = 5.5
    data['sepalWidthInput'] = 5

    data['sepalWidthInput'] = data['sepalWidthInput'].astype(float)
    
    df = pd.DataFrame(data)

    df = df.drop('Id', axis=1) 

    df['Sepal(L)'] = 5.5
    df['Sepal(W)'] = 5

    df['Species'] = df['Species'].replace([setosa, virginica, versicolor], [new_setosa,new_virginica,new_versicolor])

    df['SepalLengthCm'] = df['SepalLengthCm'].replace([old_value],new_value)    

    
```

```python
def euclid_dist(data):
    # data['Eucli_Dist'] creates new column
    data['Eucli_Dist'] = np.sqrt((data['SepalLengthCm'] - data['sepalLengthInput'])**2 + ( data['SepalWidthCm'] - data['sepalWidthInput'])**2)
    return data
```

```python
def ranking_data(data):
    data['Rank'] = data['Eucli_Dist'].rank()
    
```

```python
def euclidean_dist(row1, row2, cols,label):
    distance = 0.0

    for col in cols:
        if col != label:
            distance += np.square(row1[col] - row2[col])
    return np.sqrt(distance)
```

```python
def KNN(train, test_row,k, label):
    temp = train.copy()
    #distance measurement
    temp['dist'] = temp.apply(lambda row: euclidean_dist(row, test_row,train.columns,label), axis=1)
    #ranking
    #sorting & initial ranking of labels (setosa, virginica, versicolor) to minimum distances
    sort_distance_vals = temp['dist'].sort_values()
    k_n_distances = sort_distance_vals[:k] 
    #neighbors
    #getting majority of labels from neighbors
    # 1. temp = all temporary sorted minimum distances
    k_n = temp[temp.index.isin(k_n_distances.index)]
    # 2. list = converting all minor labels to list
    k_n_labels = list(k_n[label])
    # 3. Count all labels
    count_labels = Counter(k_n_labels)
    # 4. Predict the most common labels
    predicted_labels = count_labels.most_common()[0][0]
    return predicted_labels

# KNN(df, 0, 1, "Setosa")
```
